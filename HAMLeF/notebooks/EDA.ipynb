```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
# Load the dataset
file_path = '../data/raw/hamlef_data.csv'
data = pd.read_csv(file_path)

# Inspect the dataset
print("First 5 rows of the dataset:")
display(data.head())
```

```python
print("\nDataset info:")
data.info()
```

```python
print("\nSummary statistics:")
display(data.describe())
```

### Insights From the Summary Statistics

Hereâ€™s a breakdown of notable insights and areas for further investigation:

Key Observations:
1. Outliers:
	- 	Extreme values for CLR_BAL_AMT, AVG_CRBAL_*, and transaction amounts (*_TRAN_AMT).
	- 	Maximum values are disproportionately large compared to the 75th percentile, indicating potential outliers or data entry issues.
2. Missing Data:
	- 	Substantial missing data in columns like AVG_CRBAL_* and *_TRAN_AMT_DEBIT, with counts significantly lower than the total row count.
	- 	Missing data strategies (e.g., imputation or exclusion) need to be applied.
3. Zeros in Data:
	- 	Median values for many columns like AVG_CRBAL_* and *_TRAN_AMT_* are zero at the 25th percentile, reflecting accounts with minimal activity.
4. Class Imbalance:
	-	The PAST_DUE_DAYS column has a median value of 0, suggesting most accounts are not overdue. Further analysis is needed to confirm imbalance.
5. Transaction Distributions:
	-	Transaction counts (*_TRAN_CNT) and amounts (*_TRAN_AMT) have significant variation, with the 75th percentile markedly lower than the max values.
6. Multicollinearity:
	-	High correlation is likely between CLR_BAL_AMT, AVG_CRBAL_*, and *_TRAN_AMT. Correlation analysis will clarify relationships.

```python
# Check for missing values
print("\nMissing values:")
missing_values = data.isnull().sum().sort_values(ascending=False)
missing_percentage = (missing_values / len(data)) * 100
missing_df = pd.DataFrame({'Missing Count': missing_values, 'Percentage': missing_percentage})
display(missing_df[missing_df['Missing Count'] > 0])

# Visualize missing values
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Check for duplicates
duplicate_count = data.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_count}")
```

```python
# Column-wise unique value counts
print("\nUnique value counts in each column:")
for col in data.columns:
    print(f"{col}: {data[col].nunique()}")
```

```python
# Distribution of numeric features
numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
print("\nNumeric columns:", numeric_columns)

for col in numeric_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Distribution of categorical features
categorical_columns = data.select_dtypes(include=['object', 'category']).columns
print("\nCategorical columns:", categorical_columns)
```

```python
cat_columns = ['ACCT_STATUS', 'PAST_DUE_CATEGORY']
```

```python
for col in cat_columns:
    plt.figure(figsize=(8, 4))
    # Ensure data[col] is converted to string and drop NaN values
    sns.countplot(
        data=data, 
        x=col, 
        order=data[col].dropna().astype(str).value_counts().index, 
        palette='viridis'
    )
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Correlation analysis for numeric features
print("\nCorrelation matrix:")
correlation_matrix = data[numeric_columns].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Outlier detection using boxplots
for col in numeric_columns:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data[col], palette='Set2')
    plt.title(f'Boxplot for {col}')
    plt.show()
```

```python
# Target variable analysis (assuming "target" is the name of the column)
if 'target' in data.columns:
    print("\nTarget variable distribution:")
    plt.figure(figsize=(6, 4))
    sns.countplot(data['target'], palette='husl')
    plt.title('Target Variable Distribution')
    plt.xlabel('Target')
    plt.ylabel('Count')
    plt.show()
else:
    print("\nTarget variable not found in the dataset.")

# Pairplot for numeric features (optional for large datasets)
if len(numeric_columns) <= 5:
    print("\nPairplot for numeric features:")
    sns.pairplot(data[numeric_columns], diag_kind='kde', corner=True)
    plt.show()
```
